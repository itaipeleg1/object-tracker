```python
import cv2

tracker = cv2.TrackerCSRT_create()
cap = cv2.VideoCapture(r"C:\kela\tracker\object-tracker\data\input\MAX_0009.mp4")


ret, frame = cap.read()

height, width = frame.shape[:2]


# Resize frame for display
scale_percent = 60 
display_width = int(width * scale_percent / 100)
display_height = int(height * scale_percent / 100)

# Resize first frame for ROI selection
frame_resized = cv2.resize(frame, (display_width, display_height))
print("Select ROI and press SPACE/ENTER")

max_frames = int(30 * cap.get(cv2.CAP_PROP_FPS))
bbox = cv2.selectROI("Select Object", frame_resized, False)
cv2.destroyAllWindows()
print(f"Selected bbox (resized): {bbox}")

# Scale bbox back to original coordinates
bbox_scaled = tuple([int(b * 100 / scale_percent) for b in bbox])
print(f"Selected bbox (scaled): {bbox_scaled}")

tracker.init(frame, bbox_scaled)
frame_id = 0
fails = 0
while True:
      ret, frame = cap.read()
      if not ret or frame_id > max_frames:
          print(f"Stopped at frame {frame_id}")
          break

      success, bbox = tracker.update(frame)

      if success:
          x, y, w, h = [int(i) for i in bbox]
          center_x = x + w // 2
          center_y = y + h // 2
          radius = max(w, h) // 2
          cv2.circle(frame, (center_x, center_y), radius, (0, 255, 0), 2)
      else:
          print(f"Tracking failure at frame {frame_id}")
          break

      cv2.putText(frame, f"Frame: {frame_id}/{max_frames}", (20, 40),
                  cv2.FONT_HERSHEY_SIMPLEX, 0.75, (255, 255, 255), 2)

      # Resize for display
      frame_display = cv2.resize(frame, (display_width, display_height))
      cv2.imshow("Tracking", frame_display)

      if cv2.waitKey(30) & 0xFF == ord('q'):
          break

      frame_id += 1

cap.release()
cv2.destroyAllWindows()
print("Done!")
```
